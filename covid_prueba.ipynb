```python
ruta = "/content/drive/My Drive/Colab Notebooks/Covid-19/Datos/"
df= pd.read_csv(ruta+"Casos_Diarios_Estado_Nacional_Confirmados_210821.csv",parse_dates=True,index_col=None)
print("DataFrame Dimensiones")
print(df.shape)
df.head(3)
```

```python
### Gráfica 1
fig, ax = plt.subplots(figsize=(15,8))
#elimino la columna 'nombre' para dejar sólo datos numéricos  
# graficamos los datos localizados en .loc[fila][columna_inicial:column_final]
row = df.drop(columns=['cve_ent','poblacion','nombre']).iloc[6]['18-02-2020':'20-08-2021']
row.plot(lw=2, colormap='jet', marker='.', markersize=5,ax=ax)
#total contagios confirmados
#total_confirmados_distrito =row.sum()
#intruciones correspondientes a Estado de México
row = df.drop(columns=['cve_ent','poblacion','nombre']).iloc[14]['18-02-2020':'20-08-2021']
row.plot(lw=2, colormap='prism', marker='.', markersize=9,ax=ax)
#total_confirmados_EstMex =row.sum()
#intruciones correspondientes a Aguascalientes
row = df.drop(columns=['cve_ent','poblacion','nombre']).iloc[1]['18-02-2020':'20-08-2021']
row.plot(lw=2, colormap='magma', marker='.', markersize=9,title='Casos Confirmados COVID al día 20 de Agosto de 2021',ax=ax)
#total_confirmados_aguas =row.sum()
ax.legend(['DISTRITO FEDERAL',"MEXICO","AGUASCALIENTES"]);
```

```python
#Contagios de Mayo
row = df.drop('nombre',axis=1).iloc[6]['01-05-2021':'31-05-2021']
total_confirmados_distrito =row.sum()
row = df.drop('nombre',axis=1).iloc[14]['01-05-2021':'31-05-2021']
total_confirmados_EstMex =row.sum()
row = df.drop('nombre',axis=1).iloc[1]['01-05-2021':'31-05-2021']
total_confirmados_aguas =row.sum()
```

```python
#Contagios de Junio
row = df.drop('nombre',axis=1).iloc[6]['01-06-2021':'30-06-2021']
total_confirmados_distrito1 =row.sum()
row = df.drop('nombre',axis=1).iloc[14]['01-06-2021':'30-06-2021']
total_confirmados_EstMex1 =row.sum()
row = df.drop('nombre',axis=1).iloc[1]['01-06-2021':'30-06-2021']
total_confirmados_aguas1 =row.sum()
```

```python
#Contagios de Julio
row = df.drop('nombre',axis=1).iloc[6]['01-07-2021':'30-07-2021']
total_confirmados_distrito2=row.sum()
row = df.drop('nombre',axis=1).iloc[14]['01-07-2021':'30-07-2021']
total_confirmados_EstMex2=row.sum()
row = df.drop('nombre',axis=1).iloc[1]['01-07-2021':'30-07-2021']
total_confirmados_aguas2=row.sum()
```

```python
### Gráfica 2
total_mayo= [total_confirmados_EstMex,total_confirmados_distrito,total_confirmados_aguas]
total_junio= [total_confirmados_EstMex1,total_confirmados_distrito1,total_confirmados_aguas1]
total_julio= [total_confirmados_EstMex2,total_confirmados_distrito2,total_confirmados_aguas2]
#Numéro de gráficas de barras
N=3
ind = np.arange(N) # the x locations for the groups
width = 0.35
fig, ax = plt.subplots(figsize=(8,6))
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, total_mayo, width, color='r')
ax.bar(ind, total_junio, width,bottom=total_mayo, color='b')
ax.bar(ind, total_julio, width,bottom=total_junio, color='g')
ax.set_ylabel('Contagios')
plt.xticks(np.arange(N), ('Estado de México', 'Ciudad de México','Aguascalientes'))
ax.legend(labels=['Mayo','JUNIO','JULIO'])
plt.title('')
plt.show()
```
