## Gráficas

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.set_style("darkgrid") #Fondo gris con cuadrículas
row = data.iloc[6]['26-02-2020':'02-03-2022']
row.plot(lw=1, colormap='autumn_r', marker='.', markersize=8 , ax=ax)

row = data.iloc[32]['26-02-2020':'02-03-2022']
row.plot(lw=1, colormap='jet', marker='.', markersize=8 , ax=ax)
plt.title("Casos Diarios Confirmados",fontsize="15")
plt.ylabel("Número de casos");
ax.legend(['Ciudad de México','Nacional']);plt.grid(True)
plt.plot()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.set_style("darkgrid") #Fondo gris con cuadrículas
row = data2.iloc[6]['17-03-2020':'02-03-2022']
row.plot(lw=1, colormap='autumn_r', marker='.', markersize=8 , ax=ax)

row = data2.iloc[32]['17-03-2020':'02-03-2022']
row.plot(lw=1, colormap='jet', marker='.', markersize=8 , ax=ax)
plt.title("Nuevas Muertes Diaraias",fontsize=15)
ax.legend(['Ciudad de México','Nacional']);plt.grid(True)
plt.ylabel("Número de casos");
plt.plot()
```
